```python
from rdkit import Chem
def conf_viewer(idx):
    mol = Chem.Mol(data['mol'][0])
    mol.AddConformer(data['mol'][0].GetConformer(idx))
    return MolTo3DView(mol).show()

interact(conf_viewer, idx=ipywidgets.IntSlider(min=0,max=data['mol'][0].GetNumConformers()-1, step=1))
```
